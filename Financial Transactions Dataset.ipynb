```python
client_total = df.groupby('client_id')['amount'].sum().reset_index()

print(client_total)
```

```python
client_card_count = df.groupby('client_id')['card_id'].nunique().reset_index()
client_card_count = client_card_count.rename(columns={'card_id': 'card_count'})
print(client_card_count)
```

```python
error_count_per_client = df[df['errors'].notna()].groupby('client_id').size().reset_index(name='error_count')
print(error_count_per_client)
```

```python
client_total = client_total.merge(error_count_per_client, on='client_id', how='left')
client_total = client_total.merge(client_card_count, on='client_id', how='left')
print(client_total)
```

```python
client_total['error_count'] = client_total['error_count_x']
client_total = client_total = client_total.drop(columns=['error_count_x', 'error_count_y'])
print(client_total)
```
